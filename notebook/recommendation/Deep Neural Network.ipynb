## Import Libraries

```python
# work with data
import pandas as pd
import numpy as np

# train_test_split library
from sklearn.model_selection import train_test_split

# neural network
import tensorflow as tf

# pickle
import pickle

# import library to convert string to integer
from sklearn.preprocessing import LabelEncoder

# ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

## Import Dataset

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset.csv')
df.head()
```

## Drop Columns

```python
# drop the result column from input data ('crop')
inputs = df.drop('crop', axis = 'columns')

# save the dropped column in separate variable
outputs = df[['crop']]
```

```python
# print input
inputs.head()
```

```python
# print output
outputs.head()
```

## Convert Strings

```python
#Encoding label (i.e string format) to integer for training and testing model
label_encoder = LabelEncoder()
```

```python
# create separate column for converted values
outputs['crop_converted'] = label_encoder.fit_transform(outputs['crop'])
outputs
```

```python
# ignore warning
pd.options.mode.chained_assignment = None

# drop string column
outputs_converted = outputs.drop(['crop'], axis = 'columns')
outputs_converted
```

```python
{index : label for index, label in enumerate(label_encoder.classes_)}
```

## Split Dataset

```python
# split into 80% train and 20% test
X_train, X_test, y_train, y_test = train_test_split(inputs, outputs_converted, test_size = 0.2)
```

# Apply Neural Network

```python
a1 = tf.keras.Input(shape = (7, ))
a2 = tf.keras.layers.Dense(64, activation = 'relu')(a1)
a3 = tf.keras.layers.Dense(22, activation = 'softmax')(a2)

model = tf.keras.Model(a1,a3)

model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = 'accuracy')

result = model.fit(X_train, y_train, validation_split = 0.2, batch_size = 128, epochs = 100)
```

```python
_, accuracy = model.evaluate(X_test, y_test)
print("Accuracy of Deep Neural Network algorithm is = \033[1m", round(accuracy * 100.0, 2), "\033[0m%")
```

```python
with open('DNN_MODEL', 'wb') as f:
    pickle.dump(model, f)
```
